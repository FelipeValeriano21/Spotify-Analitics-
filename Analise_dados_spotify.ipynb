```python
import pandas as pd
import locationtagger

spotify_database = pd.read_csv("spotify_data.csv")
```

```python
spotify_database['Artist'] = spotify_database['Songs & Artist'].str.split('-').str[0]
```

```python
spotify_database['Songs & Artist'] = spotify_database['Songs & Artist'].str.split('-').str[1]
```

```python
# Agrupar por 'Artist' e somar as 'Streams'
artistas_somados = spotify_database.groupby('Artist', as_index=False)['Streams'].sum()

# Ordenar pelo total de 'Streams' e pegar os 10 artistas com mais visualizações
top_10_artists = artistas_somados.sort_values(by='Streams', ascending=False).head(10)

# Exibir os 10 artistas com mais visualizações
print(top_10_artists[['Artist', 'Streams']])

```

```python
# Agrupar por 'Artist' e somar as 'Streams'
artistas_somados = spotify_database.groupby('Artist', as_index=False)['Daily'].sum()

# Ordenar pelo total de 'Streams' e pegar os 10 artistas com mais visualizações
top_10_artists = artistas_somados.sort_values(by='Daily', ascending=False).head(10)

# Exibir os 10 artistas com mais visualizações
print(top_10_artists[['Artist', 'Daily']])
```

```python
import requests
from bs4 import BeautifulSoup
import locationtagger

def pegar_pais_wikipedia(artist_name):
    # Substitui espaços por underscores para o URL
    url = f"https://en.wikipedia.org/wiki/{artist_name.replace(' ', '_')}"
    
    # Faz a requisição para a página
    resposta = requests.get(url)
    
    # Verifica se a requisição foi bem-sucedida
    if resposta.status_code == 200:
        # Retorna o conteúdo HTML da página
        return resposta.text
    else:
        return None  # Retorna None se a requisição falhar

for artist in spotify_database['Artist']:
    # Exemplo de uso
    html = pegar_pais_wikipedia(artist)

    # Se o HTML for recuperado com sucesso, processa com BeautifulSoup
    if html:
        # Cria um objeto BeautifulSoup
        soup = BeautifulSoup(html, 'html.parser')

        # Encontra todos os parágrafos
        paragrafo = soup.find_all('p')

        # Variável para armazenar o primeiro país encontrado
        primeiro_pais = None

        # Loop para processar cada parágrafo
        for p in paragrafo:
            texto = p.get_text()

            # Extraindo entidades
            place_entity = locationtagger.find_locations(text=texto)

            # Obtendo todos os países
            if place_entity.countries:
                primeiro_pais = place_entity.countries[0]  # Captura o primeiro país encontrado
                break  # Sai do loop após encontrar o primeiro país

        # Imprime o primeiro país encontrado, se houver
        if primeiro_pais:
            print(f"O primeiro país encontrado: {primeiro_pais}")
            spotify_database['Country'] = primeiro_pais

        else:
            print("Nenhum país encontrado.")
    else:
        print("Falha ao recuperar a página.")
```
